```python
#CI_sum_prop(estimates=raw_data.loc[raw_data['Kingdom'] == 'Plants','Biomass'].values,mul_CIs=raw_data.loc[raw_data['Kingdom'] == 'Plants','Uncertainty'].values)
def integrate(x):
    x2 = x[['Biomass','Uncertainty']].dropna()
    
    return CI_sum_prop(x2['Biomass'].values,x2['Uncertainty'].values)
CI = raw_data.groupby('Kingdom').apply(integrate)
bio = raw_data.groupby('Kingdom')['Biomass'].sum()
agg_res = pd.concat([bio,CI],axis=1)
agg_res.columns = ['Biomass','Uncertainty']
agg_res.loc['fungi','Uncertainty'] = 10
agg_res.loc['bacteria','Uncertainty'] = 3
agg_res.loc['archaea','Uncertainty'] = 3
agg_res.to_csv('../results/agg_data.csv')
```
